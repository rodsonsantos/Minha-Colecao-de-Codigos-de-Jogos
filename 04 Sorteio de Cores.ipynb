```python
from random import shuffle
from time import sleep

cores = ['Vermelho', 'Rosa', 'Amarelo', 'Preto', 'Branco']

shuffle(cores)

sorteio = {
    'Clarice': cores[0],
    'Eliana': cores[1],
    'Michele': cores[2],
    'Laís': cores[3],  
    'Genilda': cores[4]
}

print('Valores sorteados:')
for k, v in sorteio.items():
    print(f'{k} tirou: {v}.')
    sleep(1)
```
